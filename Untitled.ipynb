```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split

```

```python
df = pd.read_csv('transaction_data.csv')
print(df.describe())

plt.hist(df['Amount'], bins=50)
plt.title('Transaction Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
X = df[['Amount']]  # Feature for anomaly detection
```

```python
model = IsolationForest(n_estimators=100, contamination=0.02, random_state=42)
model.fit(X)
```

```python
df['Anomaly'] = model.predict(X)
df['Anomaly'] = df['Anomaly'].apply(lambda x: 1 if x == -1 else 0)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df[df['Anomaly'] == 0]['Amount'], bins=50, alpha=0.5, label='Normal')
plt.hist(df[df['Anomaly'] == 1]['Amount'], bins=50, alpha=0.5, label='Anomaly')
plt.legend()
plt.title('Anomaly Detection in Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
comparison = df[['Amount', 'IsFraud', 'Anomaly']].groupby(['IsFraud', 'Anomaly']).size().unstack(fill_value=0)
print(comparison)
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd

# Load the data
df = pd.read_csv('transaction_data.csv')

# Use only the 'Amount' column for anomaly detection
X = df[['Amount']]

# Initialize and fit the Isolation Forest model
model = IsolationForest(n_estimators=100, contamination=0.02, random_state=42)
model.fit(X)

# Predict anomalies: -1 indicates anomaly, 1 indicates normal
df['Anomaly'] = model.predict(X)

# Convert predictions: 1 (normal) -> 0, -1 (anomaly) -> 1
df['Anomaly'] = df['Anomaly'].apply(lambda x: 1 if x == -1 else 0)

# Optionally, save the updated DataFrame to a new CSV file
df.to_csv('transaction_data_with_anomalies.csv', index=False)

# Check that the 'Anomaly' column has been added
print(df.head())
```
